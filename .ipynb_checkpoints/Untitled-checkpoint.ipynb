```python
def foo(series, name):
    print(series.quantile(1))
    for i in range(1, 9):
        print(f' to above {10 * i} % users u need {series.quantile(i / 10)} {name} or {series.quantile(i / 10) / series.quantile(1) * 100} % of maximum')
    for i in range(1, 20):
        print(f' to above {80 + i} % users u need {series.quantile((80 + i) / 100)} {name} or {series.quantile((80 + i) / 100) / series.quantile(1) * 100} % of maximum')
foo(df.Rating, 'points')
```
